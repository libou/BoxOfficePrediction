```python
import csv
import pandas as pd

d_2008 = pd.read_csv('data/IMDB_movie_detail_2008.csv')
d_2009 = pd.read_csv('data/IMDB_movie_detail_2009.csv')
d_2010 = pd.read_csv('data/IMDB_movie_detail_2010.csv')
d_2011 = pd.read_csv('data/IMDB_movie_detail_2011.csv')
d_2012 = pd.read_csv('data/IMDB_movie_detail_2012.csv')
d_2013 = pd.read_csv('data/IMDB_movie_detail_2013.csv')
d_2014 = pd.read_csv('data/IMDB_movie_detail_2014.csv')
d_2015 = pd.read_csv('data/IMDB_movie_detail_2015.csv')
d_2016 = pd.read_csv('data/IMDB_movie_detail_2016.csv')
d_2017 = pd.read_csv('data/IMDB_movie_detail_2017.csv')
d_2018 = pd.read_csv('data/IMDB_movie_detail_2018.csv')
d_2019 = pd.read_csv('data/IMDB_movie_detail_2019.csv')
```

```python
n_2008_2011 = pd.read_csv('data/IMDB_Top10000_2008_2011.csv')
n_2008 = n_2008_2011[n_2008_2011['year'] == 2008]
n_2009 = n_2008_2011[n_2008_2011['year'] == 2009]
n_2010 = n_2008_2011[n_2008_2011['year'] == 2010]
n_2011 = n_2008_2011[n_2008_2011['year'] == 2011]

n_2012_2014 = pd.read_csv('data/IMDB_Top10000_2012_2014.csv')
n_2012 = n_2012_2014[n_2012_2014['year'] == 2012]
n_2013 = n_2012_2014[n_2012_2014['year'] == 2013]
n_2014 = n_2012_2014[n_2012_2014['year'] == 2014]

n_2015_2019 = pd.read_csv('data/IMDB_Top10000_2015_2019.csv')
n_2015 = n_2015_2019[n_2015_2019['year'] == 2015]
n_2016 = n_2015_2019[n_2015_2019['year'] == 2016]
n_2018 = n_2015_2019[n_2015_2019['year'] == 2018]
n_2019 = n_2015_2019[n_2015_2019['year'] == 2019]

n_2017 = pd.read_csv('data/IMDB_Top10000_2017.csv')
```

```python
merged_2008 = n_2008.merge(d_2008, on='movie', how='outer')
merged_2008 = merged_2008[~merged_2008.content_rating.str.contains('TV')]
merged_2008.to_csv('merged_data/merged_2008.csv', index = False, encoding = 'utf_8_sig')

merged_2009 = n_2009.merge(d_2009, on='movie', how='outer')
merged_2009 = merged_2009[~merged_2009.content_rating.str.contains('TV', na=False)]
merged_2009.to_csv('merged_data/merged_2009.csv', index = False, encoding = 'utf_8_sig')

merged_2010 = n_2010.merge(d_2010, on='movie', how='outer')
merged_2010 = merged_2010[~merged_2010.content_rating.str.contains('TV')]
merged_2010.to_csv('merged_data/merged_2010.csv', index = False, encoding = 'utf_8_sig')

merged_2011 = n_2011.merge(d_2011, on='movie', how='outer')
merged_2011 = merged_2011[~merged_2011.content_rating.str.contains('TV')]
merged_2011.to_csv('merged_data/merged_2011.csv', index = False, encoding = 'utf_8_sig')

merged_2012 = n_2012.merge(d_2012, on='movie', how='outer')
# merged_2012.drop_duplicates(keep="first", inplace = True)
merged_2012.drop(merged_2012[merged_2012['gross'] == "None"].index, inplace=True)
merged_2012 = merged_2012[~merged_2012.content_rating.str.contains('TV')]
merged_2012.to_csv('merged_data/merged_2012.csv', index = False, encoding = 'utf_8_sig')

merged_2013 = n_2013.merge(d_2013, on='movie', how='outer')
merged_2013.drop(merged_2013[merged_2013['gross'] == "None"].index, inplace=True)
merged_2013 = merged_2013[~merged_2013.content_rating.str.contains('TV')]
merged_2013.to_csv('merged_data/merged_2013.csv', index = False, encoding = 'utf_8_sig')

merged_2014 = n_2014.merge(d_2014, on='movie', how='outer')
merged_2014.drop(merged_2014[merged_2014['gross'] == "None"].index, inplace=True)
merged_2014 = merged_2014[~merged_2014.content_rating.str.contains('TV', na=False)]
merged_2014.to_csv('merged_data/merged_2014.csv', index = False, encoding = 'utf_8_sig')

merged_2015 = n_2015.merge(d_2015, on='movie', how='outer')
merged_2015.drop(merged_2015[merged_2015['gross'] == "None"].index, inplace=True)
merged_2015 = merged_2015[~merged_2015.content_rating.str.contains('TV')]
merged_2015.to_csv('merged_data/merged_2015.csv', index = False, encoding = 'utf_8_sig')

merged_2016 = n_2016.merge(d_2016, on='movie', how='outer')
merged_2016.drop(merged_2016[merged_2016['gross'] == "None"].index, inplace=True)
merged_2016 = merged_2016[~merged_2016.content_rating.str.contains('TV', na=False)]
merged_2016.to_csv('merged_data/merged_2016.csv', index = False, encoding = 'utf_8_sig')

merged_2017 = n_2017.merge(d_2017, on='movie', how='outer')
merged_2017.drop(merged_2017[merged_2017['gross'] == "None"].index, inplace=True)
merged_2017 = merged_2017[~merged_2017.content_rating.str.contains('TV')]
merged_2017.to_csv('merged_data/merged_2017.csv', index = False, encoding = 'utf_8_sig')

merged_2018 = n_2018.merge(d_2018, on='movie', how='outer')
merged_2018.drop(merged_2018[merged_2018['gross'] == "None"].index, inplace=True)
merged_2018 = merged_2018[~merged_2018.content_rating.str.contains('TV', na=False)]
merged_2018.to_csv('merged_data/merged_2018.csv', index = False, encoding = 'utf_8_sig')

merged_2019 = n_2019.merge(d_2019, on='movie', how='outer')
merged_2019.drop(merged_2019[merged_2019['gross'] == "None"].index, inplace=True)
merged_2019 = merged_2019[~merged_2019.content_rating.str.contains('TV')]
merged_2019.to_csv('merged_data/merged_2019.csv', index = False, encoding = 'utf_8_sig')
```
