```python
import csv
import pandas as pd
```

```python
df = pd.DataFrame() 
for year in range(2008, 2020):
    temp = pd.read_csv('data/data_{}.csv'.format(year))
    df = pd.concat([df, temp], ignore_index=True)
```

```python
df['gross'] = df['gross'].str.replace(",", "").astype(int)
df
```

```python
directors = df['director'].tolist()

freq_directors = {}
director_gross = {}
for director in directors:
    if director in freq_directors:
        freq_directors[director] += 1
    else:
        freq_directors[director] = 1

for key in freq_directors:
    mean_gross = df[df['director'].str.contains(key, case=False)]['gross'].mean()
    director_gross[key] = mean_gross

for key in director_gross:
    df.loc[df['director'].str.contains(key, case=False), 'director_avg_gross'] = director_gross[key]

df
```

```python
countries = df['country'].tolist()

freq_countries = {}
countries_gross = {}
for country in countries:
    if country in freq_countries:
        freq_countries[country] += 1
    else:
        freq_countries[country] = 1

for key in freq_countries:
    mean_gross = df[df['country'].str.contains(key, case=False)]['gross'].mean()
    countries_gross[key] = mean_gross
for key in countries_gross:
    df.loc[df['country'].str.contains(key, case=False), 'country_avg_gross'] = countries_gross[key]

df
```
