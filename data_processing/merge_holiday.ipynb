```python
import pandas as pd
import re
import datetime
```

```python
df_holidays = pd.read_csv("data/holiday.csv")
result = []
for idx, row in df_holidays.iterrows():
    date_str = row['date']
    date = datetime.datetime.strptime(date_str, '%d-%b-%y')
    year = date.year
    result.append(year)
df_holidays['year'] = result
```

```python
# df_holidays = pd.read_csv("data/holiday.csv")
years = range(2008, 2020)

for year in years:
    df = pd.read_csv("data/data_{}.csv".format(year), keep_default_na=False, index_col=0)
    df_year = df_holidays[df_holidays['year'] == year]
    result = []
    for idx, row in df.iterrows():
        date_str = row['release_date']
        date_pattern = re.compile(r"[0-9]+ [a-zA-Z]+ [0-9]+")
        if date_pattern.match(date_str):
            date = datetime.datetime.strptime(date_str, '%d %B %Y')
        else:
            date = datetime.datetime(year, 6, 30)
        flag = False
        for _, holiday_date in df_year.iterrows():
            holiday_date = datetime.datetime.strptime(holiday_date['date'], '%d-%b-%y')
            if abs(holiday_date - date).days <= 5:
                flag = True
                break
        result.append(flag)
    df['near_holiday'] = result
    df.to_csv('data_{}.csv'.format(year), encoding='utf-8-sig')
```
