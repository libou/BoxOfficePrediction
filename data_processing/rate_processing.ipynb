```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/boxoffice_dataset.csv', index_col=0)

def rating_cleaning(x):
    if x in ['None', 'Not Rated', 'Unrated', np.NaN]:
        return 'Not Rated'
    else:
        return x

df['content_rating'] = df['content_rating'].apply(lambda x : rating_cleaning(x))
df
```

```python
df = pd.get_dummies(df, prefix=['rating'], columns=['content_rating'])
df
```
